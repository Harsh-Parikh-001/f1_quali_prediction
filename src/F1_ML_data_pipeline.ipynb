```python
def convert_to_seconds(td):
    return td.total_seconds()
```

```python
import fastf1
import pandas as pd
lst = ['FP2','FP3']
```

```python
def driver(driver_name):
  session = fastf1.get_session(2025, 2, 'FP1')
  session.load()

  lap_FP = pd.DataFrame(session.laps.pick_drivers(driver_name))
  lap_FP = lap_FP[['LapTime','SpeedI1','SpeedI2','SpeedFL','IsAccurate','Compound']]

  for i in lst:
    session = fastf1.get_session(2025, 2, i)
    session.load()
    temp = pd.DataFrame(session.laps.pick_drivers(driver_name))
    temp = temp[['LapTime','SpeedI1','SpeedI2','SpeedFL','IsAccurate','Compound']]
    lap_FP = pd.concat([lap_FP, temp], ignore_index = True)
  return lap_FP
```

```python
def data_cleaning(df):
  df['LapTime'] = df['LapTime'].apply(convert_to_seconds)

  df = df.dropna(subset=['LapTime'])

  # Quali on normal condition happens on soft tire
  df = df[df['Compound'] == 'SOFT']

  # Keep only rows where 'IsAccurate' is True
  df = df[df['IsAccurate'] == True]

  df.reset_index(drop=True, inplace=True)
  df.index = df.index + 1
  # print(df)
  return df
```

```python
def stats_add_df(df, data):
  threshold = df['LapTime'].quantile(0.65)

  top_50 = df['LapTime'][df['LapTime'] <= threshold]

  mean_val = top_50.mean()
  median_val = top_50.median()
  std_val = top_50.std()
  min_val = top_50.min()
  max_val = top_50.max()

  avg_speed_i1 = df['SpeedI1'].max()
  avg_speed_i2 = df['SpeedI2'].max()
  avg_speed_fl = df['SpeedFL'].max()

  lst = []
  lst.extend([mean_val, median_val, std_val, min_val, max_val, avg_speed_i1, avg_speed_i2, avg_speed_fl])
  data.loc[len(data)] = lst

# def stats_add_df(df, data):
#   mean_val = df['LapTime'].mean()
#   median_val = df['LapTime'].median()
#   std_val = df['LapTime'].std()
#   min_val = df['LapTime'].min()
#   max_val = df['LapTime'].max()
#   avg_speed_i1 = df['SpeedI1'].mean()
#   avg_speed_i2 = df['SpeedI2'].mean()
#   avg_speed_fl = df['SpeedFL'].mean()

#   lst = []
#   lst.extend([mean_val, median_val, std_val, min_val, max_val, avg_speed_i1, avg_speed_i2, avg_speed_fl])
#   data.loc[len(data)] = lst


# mean, meadian ,std, min, max, sppedi1, speedi2, speedfl
```
